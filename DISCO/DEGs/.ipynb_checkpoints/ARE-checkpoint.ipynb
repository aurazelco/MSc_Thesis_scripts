```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Find overlap of our sig genes to ARE

## ARE gene list

```python
ARE = pd.read_excel('data/AREsitesHuman.xlsx', header=1)
```

```python
ARE.head()
```

```python
ARE.to_csv('data/AREsitesHuman.csv')
```

```python
ARE_full = ARE['All genes containing ARE fullsites'].to_list()
```

```python
ARE_full = [item for item in ARE_full if not(pd.isna(item)) == True]
```

```python
len(ARE_full)
```

```python
ARE_half = ARE['All genes containing ARE halfsites'].to_list()
```

```python
len(ARE_half)
```

## import sex binased genes list

```python
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
from rpy2.robjects import pandas2ri

from rpy2.robjects.conversion import localconverter

# Read RRA sex-biased genes
ro.r['load']("../01A_run_RRA/output_obj/RRA_chr_label.RData")
Annot_F=ro.r["Annot_F"]
Annot_M=ro.r["Annot_M"]

RRA_F_ls=[]
RRA_M_ls=[]
brain_reg = ['AMY','CBC', 'CC', 'FC', 'HIP','MED', 'OC', 'PL','STR','TC', 'THA']
for i in range(len(brain_reg)):
    print(brain_reg[i])
    with localconverter(ro.default_converter + pandas2ri.converter):
        F_df = ro.conversion.rpy2py(Annot_F[i])
        M_df = ro.conversion.rpy2py(Annot_M[i])
    
    RRA_F_ls.append( F_df.hgnc_symbol.tolist() )
    RRA_M_ls.append( M_df.hgnc_symbol.tolist() )
```

## overlap

```python
def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))
```

```python
def intersection3(lst1, lst2, lst3):
    return list(set(lst1) & set(lst2) & set(lst3))
```

### Female biased genes overlap with ARE

```python
ARE_fh_F = [ len(intersection3(ARE_full, ARE_half, RRA)) for RRA in RRA_F_ls]
ARE_f_F = [ len(intersection(ARE_full, RRA)) for RRA in RRA_F_ls]
ARE_h_F = [ len(intersection(ARE_half, RRA)) for RRA in RRA_F_ls]
```

```python
ARE_fh_M = [ len(intersection3(ARE_full, ARE_half, RRA)) for RRA in RRA_M_ls]
ARE_f_M = [ len(intersection(ARE_full, RRA)) for RRA in RRA_M_ls]
ARE_h_M = [ len(intersection(ARE_half, RRA)) for RRA in RRA_M_ls]
```

```python
Fnum = [len(RRA) for RRA in RRA_F_ls]
Mnum = [len(RRA) for RRA in RRA_M_ls]
```

```python
not_overlap_F= []
not_overlap_M= []
for i in range(len(brain_reg)):
    not_overlap_F.append(Fnum[i]- ARE_f_F[i] - ARE_h_F[i] + ARE_fh_F[i])
    not_overlap_M.append(Mnum[i]- ARE_f_M[i] - ARE_h_M[i] + ARE_fh_M[i])
```

```python
F_df = pd.DataFrame({
    "not_overlap":not_overlap_F,
    "overlap_ARE_full_half":ARE_fh_F,
    "overlap_ARE_full":ARE_f_F,
    "overlap_ARE_half":ARE_h_F,
    }, index=brain_reg)
```

```python
plt.style.use('seaborn-dark')
```

```python
fig_size = (8, 8)
plt.figure(figsize=fig_size)
F_df.plot(kind='bar',figsize=(8,8),stacked=True)
plt.title('overlap between female biased genes and ARE')
plt.savefig('ARE_overlap.png', dpi = 150)
```

```python
M_df = pd.DataFrame({
     "not_overlap":not_overlap_M,
    "overlap_ARE_full_half":ARE_fh_M,
    "overlap_ARE_full":ARE_f_M,
    "overlap_ARE_half":ARE_h_M
    }, index=brain_reg)
```

```python
fig_size = (8, 8)
plt.figure(figsize=fig_size)
M_df.plot(kind='bar',figsize=(8,8),stacked=True)
plt.title('overlap between male biased genes and ARE')
plt.savefig('ARE_overlap.png', dpi = 150)
```

## Calculate percentage

```python
plt.rcParams.update({'font.size': 16})
```

```python
F_df_stack = F_df.apply(lambda x:x*100/sum(x),axis=1)
F_df_stack.plot(kind='bar',stacked=True, figsize=(8,8))
plt.title('% overlap between female biased genes and ARE')
plt.savefig('ARE_overlapF2.png', dpi = 150)
```

```python
M_df_stack = M_df.apply(lambda x:x*100/sum(x),axis=1)
M_df_stack.plot(kind='bar',stacked=True, figsize=(8,8))
plt.title('% overlap between male biased genes and ARE')
plt.savefig('ARE_overlapM2.png', dpi = 150)
```
