### Female

```python
brain_regions = F_df.index.to_list()
```

```python
def fisher_test(A,b,c,D):
    oddsratio, pvalue = stats.fisher_exact([[A, c], [D, b]])
    return pvalue
```

```python
background_num = 44429
X_chr_genes = 1848
Y_chr_genes = 431
```

```python
# Female-biased genes enriched in X-chromosome
# A = X_chr_genes - Gene_sig_X[i]
# c = Gene_sig_X[i]
# D = 44429 - 1848 - Gene_sig_notX[i] - Gene_sig_X[i]
```

```python
Gene_sig_X = F_df['X']
Gene_sig_notX = F_df.sum(axis = 1) - F_df['X']
```

```python
A = [X_chr_genes - Gene_sig_X[i] for i in range(len(brain_regions))]
b = Gene_sig_notX
c = Gene_sig_X
D = [44429 - 1848 - Gene_sig_notX[i] - Gene_sig_X[i] for i in range(len(brain_regions))]
```

```python
X_p_value = [fisher_test(A[i],b[i],c[i],D[i]) for i in range(len(brain_regions))]
```

```python
print(A)
print(b.to_list())
print(c.to_list())
print(D)
```

```python
F_df['X_enrich__pvalue'] = X_p_value
```

```python
F_df
```
