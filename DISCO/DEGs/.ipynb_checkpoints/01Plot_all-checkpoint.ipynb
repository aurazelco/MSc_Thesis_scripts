# Plot percentage of overlap ERE, ARE and sex-biased, brain expressed, region elevated

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', 500)
plt.style.use('seaborn-dark')
plt.rcParams.update({'font.size': 20})
```

```python
from matplotlib.patches import Patch
```

# import percentage of ARE ERE of all set 

```python
# import all percentage of ARE ERE genes for each set
F_ARE = pd.read_csv('ARE_overlap_F.csv',index_col=0)
M_ARE = pd.read_csv('ARE_overlap_M.csv',index_col= 0)
F_ERE = pd.read_csv('ERE_overlap_F.csv', index_col = 0 )
M_ERE = pd.read_csv('ERE_overlap_M.csv', index_col = 0)
bexp_ARE = pd.read_csv('brain_expressed/brain_exp_overlap_ARE.csv', index_col = 0)
bexp_ERE = pd.read_csv('brain_expressed/brain_exp_overlap_ERE.csv', index_col = 0)
reg_ARE = pd.read_csv('region_elevated/ARE_overlap_ele.csv', index_col = 0)
reg_ERE = pd.read_csv('region_elevated/ERE_overlap_ele.csv', index_col = 0)
```

```python
F_ARE.head()
```

```python
F_ERE.head()
```

## merge dataframe

```python
brain_reg = reg_ARE.index.to_list()
```

```python
reg_ARE_ls = []
reg_ERE_ls = []
for i in range(len(brain_reg)):
    reg = brain_reg[i]
    print(reg)
    df = pd.DataFrame({'Female_biased':F_ARE.loc[reg,],'Male_biased':M_ARE.loc[reg,],
                       'Brain_express':bexp_ARE.loc[reg,],
                       'Regionally_elevated':reg_ARE.loc[reg,]
                      })
    df = df.T
    df = df.reindex(columns=['not_overlap','overlap_ARE_full','overlap_ARE_full_half','overlap_ARE_half'])
    reg_ARE_ls.append(df)
    
    ERE_df = pd.DataFrame({'Female_biased':F_ERE.loc[reg,],'Male_biased':M_ERE.loc[reg,],
                           'Brain_express':bexp_ERE.loc[reg,],
                           'Regionally_elevated':reg_ERE.loc[reg,]
                          })
    reg_ERE_ls.append(ERE_df.T)
```

```python
reg_ARE_ls[0]
```

## plot ARE

```python
colors = ['grey','firebrick','orangered','orange']
legend_elements = [ Patch(facecolor='grey', edgecolor='black',
                         label='not overlap'), 
                   Patch(facecolor='firebrick', edgecolor='black',
                         label='overlap ARE fullsites'),
                   Patch(facecolor='orange', edgecolor='black',
                         label='overlap ARE halfsites')
                  ]
fig, ax = plt.subplots(2, 3,sharex='col', sharey='row', figsize = (15,10))
reg_ARE_ls[0].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'AMY',
                  ax = ax[0,0])
ax[0,0].legend(handles=legend_elements, loc='upper left')
reg_ARE_ls[1].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'CBC',
                   yticks = [0,25,50,75,100],
                  ax = ax[0,1])
reg_ARE_ls[2].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'HIP',
                  ax = ax[0,2])
reg_ARE_ls[3].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'MED',
                  ax = ax[1,0])
reg_ARE_ls[4].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'STR',
                  ax = ax[1,1])
reg_ARE_ls[5].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'THA',
                  ax = ax[1,2])
plt.tight_layout()
#fig.suptitle('Percentage of overlap of ARE genes', size = 20)
plt.savefig('ARE_6reg.png', dpi = 200)
```

## plot ERE

```python
colors = ['grey','coral']
fig, ax = plt.subplots(2, 3,sharex='col', sharey='row', figsize = (15,10))
reg_ERE_ls[0].plot(kind='bar',stacked=True, legend = True,
                  color = colors, title = 'AMY',
                  ax = ax[0,0])
reg_ERE_ls[1].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'CBC',
                  ax = ax[0,1])
reg_ERE_ls[2].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'HIP',
                  ax = ax[0,2])
reg_ERE_ls[3].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'MED',
                  ax = ax[1,0])
reg_ERE_ls[4].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'STR',
                  ax = ax[1,1])
reg_ERE_ls[5].plot(kind='bar',stacked=True, legend = False,
                  color = colors, title = 'THA',
                  ax = ax[1,2])
plt.tight_layout()
plt.savefig('ERE_6reg.png', dpi = 200)
#fig.suptitle('Percentage of overlap of ERE genes')
```
